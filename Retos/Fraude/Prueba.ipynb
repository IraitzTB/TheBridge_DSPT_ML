```python
X = df.drop(["SeriousDlqin2yrs"],axis=1)
y= df["SeriousDlqin2yrs"]
```

```python
y.value_counts(normalize=True)
```

```python
# !pip install imblearn
```

```python
from imblearn.under_sampling import RandomUnderSampler

rus = RandomUnderSampler(random_state=42)
X_resampled, y_resampled = rus.fit_resample(X, y)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_resampled,y_resampled,test_size=0.20,random_state=5)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scal = scaler.fit_transform(X_train)
X_test_scal = scaler.transform(X_test)
```

```python
# !pip install xgboost
```

```python
import xgboost
from sklearn.model_selection import RandomizedSearchCV

gbm_param_grid = {"n_estimators" :[100,150,200],
                  "max_depth":range(1,20),
                  "learning_rate":[.1,.4,.45,.5,.55,.6,],
                  "colsample_bytree":[.6,.7,.8,.9,1]
}

xgb_class = xgboost.XGBClassifier()
xgb_random = RandomizedSearchCV(param_distributions=gbm_param_grid,
                                estimator=xgb_class,scoring="roc_auc",
                                verbose = 0,
                                n_iter=100,
                                cv = 5)

xgb_random.fit(X_train_scal,y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred = xgb_random.predict(X_test_scal)
accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import roc_auc_score

y_pred = xgb_random.predict_proba(X_test_scal)[:,1]
roc_auc_score(y_test, y_pred)
```
